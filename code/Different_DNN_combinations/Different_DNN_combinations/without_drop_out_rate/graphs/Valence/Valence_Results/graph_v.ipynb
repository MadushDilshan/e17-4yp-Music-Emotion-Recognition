```python
import pandas as pd
import numpy as np
import glob
import matplotlib.pyplot as plt
```

```python


# Define the path to the directory containing the feature files
directory_path = '/content/drive/MyDrive/FYP_dataset/Different_DNN_combinations/without_drop_out_rate/graphs/Valence/Valence_Results'
# Get a list of all feature file paths in the directory
feature_files = glob.glob(directory_path + '/*.csv')

print(len(feature_files))
```

```python
df = pd.read_csv(feature_files[0])
print(df)
```

```python
epochs = []
train_set_accuracy = []
test_set_accuracy = []
train_set_mse=[]
test_set_mse=[]

epochs = df['Epochs'].values
train_set_accuracy=df['Average Train Accuracy'].values
test_set_accuracy =df['Average Test Accuracy'].values
train_set_mse=df['Train MSE'].values
test_set_mse=df['Test MSE'].values

print("epochs",epochs )
print("train_set_accuracy ",train_set_accuracy)
print("test_set_accuracy",test_set_accuracy)
print("train_set_mse",train_set_mse)
print("test_set_mse",test_set_mse)
```

```python
# Create a figure and axis
fig, ax = plt.subplots()

# Plot train_set_accuracy
ax.plot(epochs, train_set_accuracy, label='Train Set Accuracy', marker='o')

# Plot test_set_accuracy
ax.plot(epochs, test_set_accuracy, label='Test Set Accuracy', marker='x')

# Set labels and title
ax.set_xlabel('Epochs')
ax.set_ylabel('Accuracy')
ax.set_title('Train and Test Set Accuracy vs. Epochs')

# Add a legend
ax.legend()

# Show the plot
plt.grid()
plt.show()
```

```python
# Create a figure and axis
fig, ax = plt.subplots()


ax.plot(epochs, train_set_mse, label='Train Set MSE', marker='o')


ax.plot(epochs, test_set_mse, label='Test Set MSE', marker='x')

# Set labels and title
ax.set_xlabel('Epochs')
ax.set_ylabel('MSE')
ax.set_title('Train and Test Set MSE vs. Epochs')

# Add a legend
ax.legend()

# Show the plot
plt.grid()
plt.show()
```
