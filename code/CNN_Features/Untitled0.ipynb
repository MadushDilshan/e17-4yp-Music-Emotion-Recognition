```python
import numpy as np
import librosa
import librosa.display
import matplotlib.pyplot as plt
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense

from google.colab import drive
drive.mount('/content/drive')

# Load an audio file
file_path = '/content/drive/MyDrive/FYP_dataset/CNN Features/abc.wav'
y, sr = librosa.load(file_path)

# Extract Mel spectrogram using librosa
mel_spectrogram = librosa.feature.melspectrogram(y=y, sr=sr)

# Normalize the spectrogram
mel_spectrogram = librosa.power_to_db(mel_spectrogram, ref=np.max)

# Reshape the spectrogram for CNN input
mel_spectrogram = mel_spectrogram.reshape(1, mel_spectrogram.shape[0], mel_spectrogram.shape[1], 1)

# Create a simple CNN model
model = Sequential()
model.add(Conv2D(32, kernel_size=(3, 3), activation='relu', input_shape=(mel_spectrogram.shape[1], mel_spectrogram.shape[2], 1)))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Flatten())
model.add(Dense(10, activation='relu'))
model.add(Dense(5, activation='linear'))  # Change num_features based on your task

# Compile the model
model.compile(loss='mean_squared_error', optimizer='adam')

# Load pre-trained weights if available
# model.load_weights('your_pretrained_weights.h5')

# Make predictions using the model
predictions = model.predict(mel_spectrogram)

# Print the predictions
print("Predictions:", predictions)
```
