```python
import pandas as pd

# Read the arousal and valence CSV files
df_arousal = pd.read_csv('arousal.csv')
df_valence = pd.read_csv('valence.csv')

# Merge the arousal and valence DataFrames based on song_id
merged_df = pd.merge(df_arousal, df_valence, on='song_id')

# Create new columns to indicate the range category for each song
merged_df['range_category'] = 'None'

# Iterate over each row
for index, row in merged_df.iterrows():
    # Get the sample time columns
    arousal_sample_time_cols = df_arousal.columns[1:]
    valence_sample_time_cols = df_valence.columns[1:]

    # Get the arousal and valence values for the current row
    arousal_values = row[arousal_sample_time_cols].values
    valence_values = row[valence_sample_time_cols].values

    # Check the range category for the song
    if all(x > 0 for x in arousal_values) and all(x > 0 for x in valence_values):
        range_category = 'Both Arousal and Valence +'
    elif all(x < 0 for x in arousal_values) and all(x < 0 for x in valence_values):
        range_category = 'Both Arousal and Valence -'
    elif all(x > 0 for x in arousal_values) and all(x < 0 for x in valence_values):
        range_category = 'Valence + and Arousal -'
    elif all(x < 0 for x in arousal_values) and all(x > 0 for x in valence_values):
        range_category = 'Valence - and Arousal +'
    else:
        range_category = 'None'

    # Assign the range category to the current row
    merged_df.at[index, 'range_category'] = range_category

# Group the data by range category and count the occurrences for each category
range_counts = merged_df['range_category'].value_counts()

# Print the range category counts
print(range_counts)
```

```python
import pandas as pd

def find_range_of_values(song_id, start_column, end_column):
    # Read the arousal and valence CSV files
    df_arousal = pd.read_csv('arousal.csv')
    df_valence = pd.read_csv('valence.csv')

    # Check if the song_id exists in the DataFrames
    if song_id not in df_arousal['song_id'].values or song_id not in df_valence['song_id'].values:
        print("Invalid song_id",song_id)
        return

    # Extract the arousal and valence values for the specified song_id
    arousal_row = df_arousal[df_arousal['song_id'] == song_id].iloc[:, 1:]
    valence_row = df_valence[df_valence['song_id'] == song_id].iloc[:, 1:]

    # Check if arousal and valence rows exist for the specified song_id
    if arousal_row.empty or valence_row.empty:
        print("Arousal or valence values not found for the specified song_id")
        return

    # Get the columns within the specified range
    column_range = arousal_row.columns[(arousal_row.columns >= start_column) & (arousal_row.columns <= end_column)]

    # Extract the values within the column range
    arousal_values = arousal_row[column_range].values.flatten()
    valence_values = valence_row[column_range].values.flatten()

    # Check if arousal and valence values have the same number of elements
    if arousal_values.shape == valence_values.shape:
        # Find the indices where both arousal and valence have positive values
        positive_indices = (arousal_values > 0) & (valence_values > 0)
        column_names_positive = column_range[positive_indices].tolist()

        # Find the indices where both arousal and valence have negative values
        negative_indices = (arousal_values < 0) & (valence_values < 0)
        column_names_negative = column_range[negative_indices].tolist()

        # Find the indices where valence is positive and arousal is negative
        positive_valence_negative_arousal_indices = (arousal_values < 0) & (valence_values > 0)
        column_names_positive_valence_negative_arousal = column_range[positive_valence_negative_arousal_indices].tolist()

        # Find the indices where valence is negative and arousal is positive
        negative_valence_positive_arousal_indices = (arousal_values > 0) & (valence_values < 0)
        column_names_negative_valence_positive_arousal = column_range[negative_valence_positive_arousal_indices].tolist()

        # Return the column names for each range
       # Return the column names for each range as a list
        return [
            column_names_positive,
            column_names_negative,
            column_names_positive_valence_negative_arousal,
            column_names_negative_valence_positive_arousal
        ]
    else:
        print("Arousal and valence values have different shapes")
        return


# Example usage
song_id = 3
start_column = 'sample_15000ms'
end_column = 'sample_18000ms'

result = find_range_of_values(song_id, start_column, end_column)
if result is not None:
    print("Column names for positive valence and positive arousal:")
    print(result)

```
