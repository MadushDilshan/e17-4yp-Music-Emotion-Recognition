```python
"""
Cleans the dataset so that it can be used for machine learning algorithms
"""
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

# Handling Missing Values -AV files

```python
# a function to handle null values in AV files
def hadndling_missing_values_arosal_and_valence(url):

  url='https://drive.google.com/uc?id=' + url.split('/')[-2]
  data_frame=pd.read_csv(url)
  rows,columns=data_frame.shape
  print(" ")
  print("rows , columns : ",rows,columns)

  print("            before pre-prosessing")
  print(data_frame.isnull().sum())

  result_data_frame=data_frame.dropna(axis=1) #drop all columns with missing values
  print(" ")
  print("            after pre-prosessing ")
  print(result_data_frame.isnull().sum())

  print(" ")
  return result_data_frame
```

```python
# checking for song ids and therir order in A,V files

dynamic_valence_data=hadndling_missing_values_arosal_and_valence('https://drive.google.com/file/d/1uP1MHWFkf580mYAKHyOzbMaPDBAnl0C4/view?usp=sharing')
dynamic_arosal_data=hadndling_missing_values_arosal_and_valence('https://drive.google.com/file/d/1dXei8wKz-lbWA0AGWWJhVgX7sdtoKhUZ/view?usp=sharing')

valence_songIDs=list(dynamic_valence_data['song_id'])
print("valence songIDS : ",valence_songIDs)

arosal_songIDs=list(dynamic_arosal_data['song_id'])
print("arosal songIDS  : ",arosal_songIDs)

correct_order=True

if(len(valence_songIDs)==len(arosal_songIDs)):

  for index in range(len(valence_songIDs)):
    if(valence_songIDs[index]!=arosal_songIDs[index]):
      print("song id is out of order")
      correct_order=False
      break

  if(correct_order==True):
    print("songs id order is correct")

else:
  print("song ids are incorrect")

```

```python
 # checking for the column name order in AV files

valence_column_names=list(dynamic_valence_data)  # returns a list of all of the column names in the DataFrame, in the order in which they appear.
arosal_column_names=list(dynamic_arosal_data)

print("valence columns :",valence_column_names)
print("arosal columns  :",arosal_column_names)

if(valence_column_names==arosal_column_names):
  print("same column names")
```

# Handling Outliers - AV files

```python
# Checking for outliers in AV files
# AV values should be in the range (-1,1)

valence_columns = dynamic_valence_data.columns[1:]  # exclude song id column
valence_values = dynamic_valence_data[valence_columns].values.flatten() # get all valence values

arosal_columns=dynamic_arosal_data.columns[1:]
arosal_values=dynamic_arosal_data[arosal_columns].values.flatten() # get all arosal values

fig, axes = plt.subplots(1, 2, figsize=(12, 4))

# Plot the histogram
sns.histplot(data=valence_values, ax=axes[0], kde=True)
axes[0].set_title('Valence Distribution')


# Plot the histogram
sns.histplot(data=arosal_values, ax=axes[1], kde=True)
axes[1].set_title('Arosal Distribution')
```

as all the AV values are in the range (-1,1), there is no need of handling outliers for the AV files
