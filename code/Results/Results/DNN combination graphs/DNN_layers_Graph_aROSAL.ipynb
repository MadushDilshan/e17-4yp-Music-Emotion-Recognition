```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load data from the CSV file
from google.colab import drive
drive.mount('/content/drive')

# Define the path to the directory containing the feature files
directory_path = '/content/drive/MyDrive/FYP_dataset/features'
results_file = '/content/drive/MyDrive/FYP_dataset/Different_DNN_combinations/results_dnn_differnt_layers - Sheet1(1).csv'
df = pd.read_csv(results_file)

# Convert DNN configurations to strings
dnn_configurations = df['layres and units'].astype(str)
accuracies = df['Accuracy - V'] * 100  # Convert to percentages
accuracies = accuracies.round(2)  # Round to two decimal places
precisions = df['Precision - V'] * 100  # Convert to percentages
precisions = precisions.round(2)  # Round to two decimal places
recalls = df['Recall - V'] * 100  # Convert to percentages
recalls = recalls.round(2)  # Round to two decimal places

# Create a bar plot
plt.figure(figsize=(10, 10))  # Adjust the figure size as needed

# Set a custom y-axis range to visually see the differences
plt.ylim(60, 95)  # Adjust the limits as needed

bar_width = 0.3
index = np.arange(len(dnn_configurations))
bar1 = plt.bar(index, accuracies, bar_width, color='blue', label='Accuracy')
bar2 = plt.bar(index + bar_width, precisions, bar_width, color='green', label='Precision')
bar3 = plt.bar(index + 2 * bar_width, recalls, bar_width, color='red', label='Recall')

plt.xlabel('DNN Configuration')
plt.ylabel('Metrics (%)')
plt.title('Performance Metrics for Different DNN Configurations')
plt.xticks(index + bar_width, dnn_configurations, rotation=90)  # Rotate x-axis labels for readability
plt.legend()

# Annotate the bars with their values at the top
for bars in [bar1, bar2, bar3]:
    for bar, value in zip(bars, accuracies if bars[0].get_facecolor() == 'b' else (precisions if bars[0].get_facecolor() == 'g' else recalls)):
        plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height() + 0.1, f'{value}%', ha='center', va='bottom', fontsize=9)

# Display the plot or save it to a file
plt.tight_layout()
plt.show()  # To display the plot in your Jupyter Notebook or script
# To save the plot as an image file, you can use plt.savefig('metrics_plot.png')
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load data from the CSV file
from google.colab import drive
drive.mount('/content/drive')

# Define the path to the directory containing the feature files
directory_path = '/content/drive/MyDrive/FYP_dataset/features'
results_file = '/content/drive/MyDrive/FYP_dataset/Different_DNN_combinations/results_dnn_differnt_layers - Sheet1(1).csv'
df = pd.read_csv(results_file)

# Convert DNN configurations to strings
dnn_configurations = df['layres and units'].astype(str)

# Extract the metrics you want to display
metrics = ['R2 Score - A', 'F1 Score - A', 'Accuracy - A', 'Precision - A', 'Recall - A']

# Create a bar plot
plt.figure(figsize=(16, 10))  # Adjust the figure size as needed

# Set a custom y-axis range to visually see the differences
plt.ylim(60, 92)  # Adjust the limits as needed

bar_width = 0.15
index = np.arange(len(dnn_configurations))

# Plot each metric as a bar
for i, metric in enumerate(metrics):
    values = df[metric] * 100
    values = values.round(2)
    plt.bar(index + (i * bar_width), values, bar_width, label=metric)

plt.xlabel('DNN Configuration')
plt.ylabel('Values')
plt.title('Performance Metrics for Different DNN Configurations')
plt.xticks(index + bar_width * 2, dnn_configurations, rotation=90)  # Rotate x-axis labels for readability

# Annotate the bars with their values just above the bars
for i, metric in enumerate(metrics):
    values = df[metric] * 100
    values = values.round(2)
    for j, value in enumerate(values):
        plt.text(index[j] + (i * bar_width), value + 0.5, f'{value}', ha='center', va='bottom', fontsize=9)

# Move the legend to the bottom
plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.4), fancybox=True, shadow=True, ncol=len(metrics))

# Display the plot or save it to a file
plt.tight_layout()
plt.show()  # To display the plot in your Jupyter Notebook or script
# To save the plot as an image file, you can use plt.savefig('metrics_bar_chart.png')
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load data from the CSV file
from google.colab import drive
drive.mount('/content/drive')

# Define the path to the directory containing the feature files
directory_path = '/content/drive/MyDrive/FYP_dataset/features'
results_file = '/content/drive/MyDrive/FYP_dataset/Different_DNN_combinations/results_dnn_differnt_layers - Sheet1(1) .csv'
df = pd.read_csv(results_file)

# Convert DNN configurations to strings
dnn_configurations = df['layres and units'].astype(str)
accuracies = df['Accuracy - A'] * 100  # Convert to percentages
accuracies = accuracies.round(2)  # Round to two decimal places

# Add RMSE and Recall - V values
rmse_values = df['RMSE - A'] * 100  # Assuming the column name is 'RMSE - V'
recall_values = df['R2 Score - A'] * 100  # Assuming the column name is 'Recall - V'
rmse_values = rmse_values.round(2)  # Round to two decimal places
recall_values = recall_values.round(2)  # Round to two decimal places

# Create a bar plot
plt.figure(figsize=(12, 6))  # Adjust the figure size as needed

# Set a custom y-axis range to visually see the differences
plt.ylim(0, 100)  # Adjust the limits as needed

# Create bars for accuracy, RMSE, and Recall - A
bar_width = 0.28
index = np.arange(len(dnn_configurations))
plt.bar(index, accuracies, bar_width, label='Accuracy -A', color='blue', align='center')
plt.bar(index + bar_width, rmse_values, bar_width, label='RMSE-A', color='red', alpha=0.7, align='center')
plt.bar(index + 2 * bar_width, recall_values, bar_width, label='R2 Score - A', color='green', alpha=0.7, align='center')

plt.xlabel('DNN Configuration')
plt.ylabel('Values')
plt.title('Performance Metrics for Different DNN Configurations for Arousal')
plt.xticks(index + bar_width, dnn_configurations, rotation=90)  # Rotate x-axis labels for readability

# Annotate the bars with their values at the center of the bars
for i, (acc, rmse, recall) in enumerate(zip(accuracies, rmse_values, recall_values)):
    plt.text(index[i], acc + 2, f'{acc}%', ha='center', va='bottom', fontsize=9)
    plt.text(index[i] + bar_width, rmse + 2, f'{rmse}%', ha='center', va='bottom', fontsize=9)
    plt.text(index[i] + 2 * bar_width, recall + 2, f'{recall}%', ha='center', va='bottom', fontsize=9)

# Move the legend outside the graph area
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))

# Display the plot or save it to a file
plt.tight_layout()
plt.show()  # To display the plot in your Jupyter Notebook or script
# To save the plot as an image file, you can use plt.savefig('metrics_plot.png')
```
