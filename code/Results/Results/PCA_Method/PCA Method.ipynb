```python
import pandas as pd
import numpy as np
import glob
import matplotlib.pyplot as plt
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
# Define the path to the directory containing the feature files
directory_path = '/content/drive/MyDrive/FYP_dataset/Results/PCA_Method/PCA_Method_Data_Selection-Sheet1.csv'

# Get a list of all feature file paths in the directory
feature_files = glob.glob(directory_path)

print(len(feature_files))
```

```python
df = pd.read_csv(feature_files[0])
print(df)
```

```python
NOC = df['Number of Components'].values
Explained_Arrousal = df['Explained Varience Arrousal'].values
Explained_Varience =df['Explained Varience Valence'].values

print("NOC",NOC )
print("Explained_Arrousal ",Explained_Arrousal)
print("Explained_Varience",Explained_Varience)
```

```python
# Create a figure and axis
fig, ax = plt.subplots()

# Plot train_set_accuracy
ax.plot(NOC, Explained_Arrousal, label='Explained Varience Arrousal', marker='o')

# Plot test_set_accuracy
ax.plot(NOC, Explained_Varience, label='Explained Varience Valence', marker='x')

# Set labels and title
ax.set_xlabel('Number Of Components')
ax.set_ylabel('Explained Varience')
ax.set_title('Number of Components Vs Explained Varience')

# Add a legend
ax.legend()

# Show the plot
plt.grid()
plt.show()
```
