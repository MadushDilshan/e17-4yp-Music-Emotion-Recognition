```python
# Define the path to the directory containing the feature files
import pandas as pd
import numpy as np
import glob
import matplotlib.pyplot as plt
directory_path = '/content/drive/MyDrive/FYP_dataset/Results/Comparison_CFS_AND_PCA/Comparison-Sheet1.csv'

# Get a list of all feature file paths in the directory
feature_files = glob.glob(directory_path)

print(len(feature_files))
```

```python
df = pd.read_csv(feature_files[0])
print(df)
```

```python
Arrosal_Accuracy =  df['Arrousal Accuracy'].values
Valence_Accuracy = df['Valence Accuracy'].values

print("Arrosal_Accuracy",Arrosal_Accuracy )
print("Valence_Accuracy ",Valence_Accuracy)
```

```python
# set height of bar
CFS = []
PCA = []
properties = [Arrosal_Accuracy, Valence_Accuracy]
for i in range(0,len(properties),1):
  CFS.append((properties[i][0]*100).round(2))
for i in range(0,len(properties),1):
  PCA.append((properties[i][1]*100).round(2))

print(CFS)
print(PCA)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Set width of bar
barWidth = 0.35  # Decrease the width to reduce the gap

# Assuming you have defined CFS and PCA arrays


# Set position of bar on X axis
br1 = np.arange(len(CFS))
br2 = [x + barWidth for x in br1]

# Make the plot
plt.bar(br1, CFS, color='darkblue', width=barWidth, edgecolor='grey', label='CFS')
plt.bar(br2, PCA, color='r', width=barWidth, edgecolor='grey', label='PCA')

# Add accuracy values on top of the bars
for i, value in enumerate(CFS):
    plt.text(br1[i] , value + 0.001, f'{value:.2f}'+"%", ha='center', va='bottom', fontsize=10,  color='black')

for i, value in enumerate(PCA):
    plt.text(br2[i], value + 0.001, f'{value:.2f}' + "%", ha='center', va='bottom', fontsize=10, color='black')


# Adding Xticks
plt.xlabel('Property', fontweight='bold', fontsize=15)
plt.ylabel('Accuracy (%)', fontweight='bold', fontsize=15)
plt.title("Accuracy for different feature selection methods")
plt.xticks([(r1 + r2) / 2 for r1, r2 in zip(br1, br2)], ['Arrousal', 'Valence'])

plt.legend(loc="upper right", bbox_to_anchor=(1,1),borderaxespad=-6)

#plt.legend()
plt.show()
```
