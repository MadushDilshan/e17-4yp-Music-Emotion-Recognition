```python
import pandas as pd
import numpy as np
import glob
import matplotlib.pyplot as plt
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
# Define the path to the directory containing the feature files
directory_path = '/content/drive/MyDrive/FYP_dataset/Results/CFS_Method/CFS_Method_Data_Selection_Sheet1.csv'

# Get a list of all feature file paths in the directory
feature_files = glob.glob(directory_path)

print(len(feature_files))
```

```python
df = pd.read_csv(feature_files[0])
print(df)
```

```python
Threshold_val = df['Threshold Value'].values
Arrousal_Accuracy=df['Arrousal Accuracy'].values
Valence_Accuracy =df['Valence Accuracy'].values

print("Threshold_val",Threshold_val )
print("Arrousal_Accuracy ",Arrousal_Accuracy)
print("Valence_Accuracy",Valence_Accuracy)
```

```python
# Create a figure and axis
fig, ax = plt.subplots()

# Plot train_set_accuracy
ax.plot(Threshold_val, Arrousal_Accuracy, label='Arrousal Accuracy', marker='o')

# Plot test_set_accuracy
ax.plot(Threshold_val, Valence_Accuracy, label='Valence Accuracy', marker='x')

# Set labels and title
ax.set_xlabel('Threshold Value')
ax.set_ylabel('Accuracy')
ax.set_title('Threshold Value Vs Model Accuracy')

# Add a legend
ax.legend()

# Show the plot
plt.grid()
plt.show()
```
