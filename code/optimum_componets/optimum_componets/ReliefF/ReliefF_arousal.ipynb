```python
import pandas as pd
import numpy as np
import glob
import matplotlib.pyplot as plt
from skrebate import ReliefF
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
# Define the path to the directory containing the feature files
directory_path = '/content/drive/MyDrive/FYP_dataset/features'
```

```python
# Get a list of all feature file paths in the directory
feature_files = glob.glob(directory_path + '/*.csv')
```

```python
print(len(feature_files))
```

```python
df= pd.read_csv(feature_files[0])
```

```python
for i in range(1,len(feature_files)):
  df_i=pd.read_csv(feature_files[i])
  df = pd.concat([df, df_i], ignore_index=True)
```

```python
rows,cols=df.shape
print("rows",rows)
print("cols",cols)
```

```python
df.head(10)
```

```python
arosal_dataset=df.drop(['Valence_Value', 'frameTime'], axis=1)
rows,cols=arosal_dataset.shape
print("rows",rows)
print("cols",cols)
arosal_dataset.head(5)
```

```python
arosal_dataset_X=arosal_dataset.drop(['Arousal_Value'], axis=1)
arosal_dataset_y=arosal_dataset['Arousal_Value']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(arosal_dataset_X, arosal_dataset_y, test_size=0.2, random_state=42)
```

```python
rrelieff = ReliefF()
rrelieff.fit(X_train.values, y_train.values)
```

```python
feature_scores = rrelieff.feature_importances_
```

```python
num_selected_features = 10  # Change this value according to your preference
selected_feature_indices = np.argsort(feature_scores)[::-1][:num_selected_features]

selected_features = arosal_dataset_X.columns[selected_feature_indices]
print("Selected features:", selected_features)
```
