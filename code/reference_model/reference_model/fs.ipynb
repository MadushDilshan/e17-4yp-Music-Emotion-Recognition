```python
arosal_dataset=df.drop(['Valence_Value', 'frameTime'], axis=1)
rows,cols=arosal_dataset.shape
print("rows",rows)
print("cols",cols)
arosal_dataset.head(5)
```

```python
print(arosal_dataset)
```

```python
def correlation_feature_selection(data, target_variable, threshold=0.5):
    # Compute the correlation matrix
    corr_matrix = data.corr()

    # Get the absolute correlation values with the target variable
    corr_with_target = abs(corr_matrix[target_variable])

    # Select features above the threshold
    selected_features = corr_with_target[corr_with_target > threshold].index.tolist()

    return selected_features
```

```python
 # Set the threshold for correlation. Features with correlation greater than this value will be selected.
threshold_value = 0.15

# Perform correlation-based feature selection
selected_features = correlation_feature_selection(arosal_dataset,'Arousal_Value', threshold_value)

print("Selected features:", selected_features)
print(len(selected_features))
```

```python
feature_selected_df = df[selected_features]
rows,cols = feature_selected_df.shape
print(rows)
print(cols)
```
